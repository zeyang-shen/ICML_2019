```python
def generate_init_weights(motif_list, motif_len=19, motif_dict=motif_dict):
    initial_weights = []
    for im in motif_list:
        current_motif = motif_dict[im]
#         w = np.array(pd.DataFrame(current_motif.pssm).T)
        w = np.array(np.log2((pd.DataFrame(motif_dict[im].pwm).T+0.01)/0.25))
        rdm = np.zeros((4,motif_len-w.shape[1]))
        half_loc = int((motif_len-w.shape[1])/2)
        w = np.concatenate([rdm[:,:half_loc], w, rdm[:,half_loc:]], axis=1)
        initial_weights.append(np.array(w))
    initial_weights = np.array(initial_weights).T
    return initial_weights
```

# Build Model

```python
def CNN_model(num_class=2, input_shape=(200, 20)):
    # define input tensors
    inputs = Input(shape=input_shape, name='input')

    # 1st Conv
    conv1_act = ReLU(name='conv1_act')
    maxpool1 = MaxPooling1D(pool_size=5, name='maxpool1')
    batchNorm_1 = BatchNormalization(name='batchNorm1')
    conv1_out = batchNorm_1(maxpool1(conv1_act(inputs)))
    
    # 2nd Conv
    conv2 = Conv1D(15, kernel_size=10, 
                   padding='same', use_bias=False, name='conv2')
    conv2_act = ReLU(name='conv2_act')
    maxpool2 = MaxPooling1D(pool_size=5, name='maxpool2')
    batchNorm_2 = BatchNormalization(name='batchNorm2')
    
    conv2_out = batchNorm_2(maxpool2(conv2_act(conv2(conv1_out))))
    
    # 1st Fully-connected
    flatten_layer = Flatten(name='flat') # change 2-D array into 1-D
    dropout0 = Dropout(0.3, name='dropout0')
    flat = dropout0(flatten_layer(conv2_out))
    
    dense1 = Dense(12, use_bias=True, name='dense1')
    dense1_act = ReLU(name='dense1_act')
    dropout1 = Dropout(0.3, name='dropout1')
    
    fc_out = dropout1(dense1_act(dense1(flat)))
    
    # Final classification
    if num_class == 1:
        dense3 = Dense(num_class, use_bias=True, name='dense3')
        outputs = dense3(fc_out)
    else:
        dense3 = Dense(num_class, use_bias=True, name='dense3')
        dense3_act = Softmax(name='dense3_act')
        outputs = dense3_act(dense3(fc_out))
    
    model = Model(inputs=inputs, outputs=outputs)
    model.summary()
    
    return model
```

```python
model = CNN_model(num_class=2)
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

## Compute motif score matrix

```python
def padding(inp, target_len, value=0):
    if target_len < len(inp):
        return inp[:target_len]
    
    padding_len = target_len - len(inp)
    if value == None: # assume random values
        padded = list(np.random.randn(padding_len//2)) + list(inp) + list(np.random.randn(padding_len-padding_len//2))
    else:
#         padded = list([value]*(padding_len//2)) + list(inp) + list([value]*((padding_len-padding_len//2)))
        padded = list(inp) + list([value]*padding_len)
    
    return np.array(padded)

def compute_scores(motif_dict, motif, seq_list):
    motif_object = motif_dict[motif]
    motif_object.pseudocounts = 0.01
    fwd_pssm = motif_object.pssm
    rev_pssm = fwd_pssm.reverse_complement()
    
    all_scores = []
    for seq in seq_list:
        fwd_scores = fwd_pssm.calculate(seq) # scores for forward orientation
        rev_scores = rev_pssm.calculate(seq) # scores for reverse orientation
        scores = np.max([fwd_scores, rev_scores], axis=0)
        padded_scores = padding(scores, 200, value=0)
        if padded_scores is None:
            print('Padding Failed!')
            return
        all_scores.append(padded_scores)
    all_scores = np.array(all_scores)
            
    return motif, all_scores


def compute_scores_parallel(motif_dict, motif_list, seq_list, p=1):
    '''
    Compute motif scores in parallel
    '''
    alphabet = Bio.Seq.IUPAC.Alphabet.IUPAC.IUPACUnambiguousDNA()
    seq_list = [Seq(s, alphabet) for s in seq_list]
    
    pool = mp.Pool(processes=p)
    results = [pool.apply_async(compute_scores, args=(motif_dict, motif, seq_list)) for motif in motif_list]
    results = [p.get() for p in results]
    
    # Combine results from multiple processors
    motif_score_array = []
    for r in results:
        if r is None:
            print('Failed!')
            return
        motif, scores = r
        motif_score_array.append(scores)
    
    return np.array(motif_score_array)
```

```python
motif_score_array = compute_scores_parallel(motif_dict, top_motifs, seqs, 10)

input_array = np.array([motif_score_array[:,i].T for i in range(len(seqs))])
input_array = np.nan_to_num(input_array)
input_array.shape
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(input_array, labels, test_size=0.2)
print(x_train.shape)
```

# Training

```python
class Histories(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.aucs = []
        self.losses = []

    def on_epoch_end(self, epoch, logs={}):
        self.losses.append(logs.get('loss'))
        y_pred = self.model.predict(x_test)
        self.aucs.append(metrics.roc_auc_score(y_test, y_pred))
        return
```

```python
batch_size = 512
epochs = 40
history = Histories()
trials = 1
all_aucs = []
for t in range(trials):
    model.fit(x_train, y_train,
              batch_size=batch_size,
              epochs=epochs,
              verbose=2,
              validation_data=(x_test, y_test), 
              callbacks=[history])

    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('AUC:', history.aucs[-1])
    all_aucs.append(history.aucs[-1])

# Save trained model
# model_file = './CNN_model_ATAC_notx.h5'
# model.save(model_file, overwrite=True)
```

# Gradients

```python
def get_pre_activation_output(layer, output):
    new_output = []
    if hasattr(layer, 'activation') and not layer.activation == keras.activations.linear:
        for inp in output.op.inputs:
            new_output.append(inp)
        return new_output
    else:
        return [output]
```

```python
layer_name = 'dense3'
selected_layer = model.get_layer(layer_name)
sel_output = get_pre_activation_output(selected_layer, selected_layer.output)[0]
inp = model.inputs
saliency = K.gradients(sel_output, inp)
gradient_function = K.function(inp, saliency)
```

```python
score_min = np.array([motif_dict[m].pssm.min for m in top_motifs])
score_max = np.array([motif_dict[m].pssm.max for m in top_motifs])
scaled_sc = (sc[0]-0)/(score_max-0)
```

```python
grad_score = gradient_function([sc])[0]
scoreXinput = grad_score * (scaled_sc)#(scaled_sc, sc, sc>0)
val = np.abs(scoreXinput[0])
val = val/val.max()
```

```python
all_combs = []
all_m = []
all_motif_max_vals = []
c = np.random.choice(len(seqs)//2, 20000, replace=False)
# c = np.arange(20000)
indices = []
for n in range(len(c)):
    if n % 1000 == 0:
        print('Processing:', n)
    ind = c[n]
    mScores = input_array[ind:ind+1]
    scaled_mScores = (mScores[0]-0)/(score_max-0)
    grad_score = gradient_function([mScores])[0]
    if (grad_score != 0).sum() == 0:
        continue
    scoreXinput = grad_score*(scaled_mScores)#(mScores > 0, scaled_mScores) # scaled motif scores work best!
    val = np.abs(scoreXinput[0])
    val = val/val.max()
    indices.append(ind)
    
    top_list = []
    pos_max_val = val.max(axis=1)
    motif_max_val = (val*(val == val.max(axis=1, keepdims=1))).max(axis=0)
    all_motif_max_vals.append(motif_max_val)
    for i in np.argsort(pos_max_val)[::-1]:
        i_m = np.array(top_motifs)[np.abs(val).argmax(axis=1)[i]]
        if i_m in top_list:
            continue
        top_list.append(i_m)
        if len(top_list) >= 2:
            break
    all_combs.append(' + '.join(np.sort(top_list)))
    all_m.extend(top_list)
    
all_combs = np.array(all_combs)
```
